```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None) 
pd.set_option('display.max_columns', None) 
pd.set_option('display.width', None) 
pd.set_option('display.max_colwidth', None)
```

```python
property_df = pd.read_csv("HomeHarvest_Chicago_10years.csv")
```

# Data Dictionary


### Basic Information

-   property_url: The URL link to the property's online listing or webpage.
-   mls: Multiple Listing Service number, a unique identifier for properties listed on the MLS system.
-   mls_id: Another identifier used within the MLS system, often specific to the region or listing service.
-   status: Current status of the property (e.g., "For Sale", "Sold", "Under Contract").

### Address Details

-   street: The street address of the property.
-   unit: If applicable, the unit number of the property (used for multi-unit buildings like apartments or condos).
-   city: The city where the property is located.
-   state: The state or province where the property is located.
-   zip_code: The postal or ZIP code for the property's location.

### Property Description

-   style: Architectural style of the property (e.g., ranch, colonial, contemporary).
-   beds: Number of bedrooms in the property.
-   full_baths: Number of full bathrooms (with a shower or bathtub).
-   half_baths: Number of half bathrooms (toilet and sink only).
-   sqft: Total square footage of the property.
-   year_built: The year in which the property was built.
-   stories: Number of floors or levels the property has.
-   lot_sqft: Size of the property's lot in square feet.

### Property Listing Details

-   days_on_mls: Number of days the property has been listed on the MLS.
-   list_price: The price at which the property is currently listed.
-   list_date: The date when the property was listed on the market.
-   sold_price: The price at which the property was most recently sold.
-   last_sold_date: The date when the property was last sold.
-   price_per_sqft: The list price of the property divided by its square footage.
-   hoa_fee: Homeowners Association fee, if applicable.

### Location Details

-   latitude: The geographical latitude of the property.
-   longitude: The geographical longitude of the property.

### Parking Details

-   parking_garage: Indicates the presence and details of a parking garage (e.g., number of spaces, attached or detached).

# Univariate Analysis

```python
numerical_vars = ['beds', 'sqft', 'year_built', 'days_on_mls', 'sold_price', 
                  'lot_sqft', 'price_per_sqft', 'latitude', 'longitude', 
                  'stories', 'hoa_fee']

print("Summary Statistics for Numerical Variables:")
print(property_df[numerical_vars].describe())

for var in numerical_vars:
    plt.figure(figsize=(8, 4))
    sns.histplot(property_df[var], bins=20, kde=True)
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_vars = ['style', 'community', 'neighborhood']

for var in categorical_vars:
    print(f"Frequency of Categories in {var}:")
    print(property_df[var].value_counts())

    plt.figure(figsize=(12, 6))
    sns.countplot(data=property_df, x=var, order=property_df[var].value_counts().index)
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.xticks(rotation=90)
    plt.show()
```

# Property Market Overview

## Total number of properties

```python
property_df['last_sold_date'] = pd.to_datetime(property_df['last_sold_date'])

property_df['sale_year'] = property_df['last_sold_date'].dt.year

properties_sold_each_year = property_df.groupby('sale_year').size().reset_index(name='total_properties_sold')

plt.figure(figsize=(12, 6))
plt.bar(properties_sold_each_year['sale_year'], properties_sold_each_year['total_properties_sold'], color='blue')
plt.title('Total Number of Properties Sold Each Year')
plt.xlabel('Year')
plt.ylabel('Total Properties Sold')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

## Distribution of property types

```python
property_type_counts = property_df.groupby('neighborhood')['style'].value_counts().unstack(fill_value=0)

property_type_counts.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('Distribution of Property Types by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Number of Properties')
plt.xticks(rotation=90)  
plt.legend(title='Property Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()  
plt.show()
```

```python
property_df['last_sold_date'] = pd.to_datetime(property_df['last_sold_date'])

property_df['sale_year'] = property_df['last_sold_date'].dt.year

neighborhood_group_results = {}
for group_name, group_data in property_df.groupby('neighborhood'):
    properties_sold_each_year = group_data.groupby('sale_year').size().reset_index(name='total_properties_sold')
    neighborhood_group_results[group_name] = properties_sold_each_year

num_groups = len(neighborhood_group_results)
fig, axs = plt.subplots(num_groups, 1, figsize=(12, 6 * num_groups), sharex=True)

for i, (group_name, result_df) in enumerate(neighborhood_group_results.items()):
    axs[i].bar(result_df['sale_year'], result_df['total_properties_sold'], color='blue')
    axs[i].set_title(f'Total Properties Sold Each Year in {group_name}')
    axs[i].set_ylabel('Total Properties Sold')
    axs[i].grid(axis='y', linestyle='--', alpha=0.7)
    axs[i].tick_params(axis='x', labelrotation=45)
    
    for x, y in zip(result_df['sale_year'], result_df['total_properties_sold']):
        axs[i].annotate(str(y), (x, y), textcoords="offset points", xytext=(0, 5), ha='center')
    
plt.xlabel('Year')
plt.tight_layout(pad=3.0)
plt.show()
```

### Average, median, range of sold prices per year

```python
property_df['sale_year'] = property_df['last_sold_date'].dt.year

grouped_df = property_df.groupby(['neighborhood', 'sale_year'])

stats_df = grouped_df['sold_price'].agg(['mean', 'median', lambda x: x.max() - x.min()])
stats_df.rename(columns={'mean': 'Average Sold Price', 'median': 'Median Sold Price', '<lambda_0>': 'Range of Sold Prices'}, inplace=True)

stats_df['Average Sold Price'] = stats_df['Average Sold Price'].apply(lambda x: f"${x:,.2f}")
stats_df['Median Sold Price'] = stats_df['Median Sold Price'].apply(lambda x: f"${x:,.2f}")
stats_df['Range of Sold Prices'] = stats_df['Range of Sold Prices'].apply(lambda x: f"${x:,.2f}")

stats_df.head(60)
```

### Yearly average and median price trends

```python
grouped_data = property_df.groupby(['neighborhood', 'sale_year'])['sold_price'].agg(['mean', 'median']).reset_index()

neighborhood_groups = grouped_data['neighborhood'].unique()

for group in neighborhood_groups:
    group_data = grouped_data[grouped_data['neighborhood'] == group]
    plt.figure(figsize=(12, 6))
    sns.set_palette("husl")  

    plt.plot(group_data['sale_year'], group_data['mean'], label=f'{group} - Mean', marker='o')
    
    plt.plot(group_data['sale_year'], group_data['median'], label=f'{group} - Median', marker='x')
    
    plt.xlabel('Year')
    plt.ylabel('Price')
    plt.title(f'Yearly Average and Median Price Trends for {group} Neighborhood Group')
    plt.xticks(group_data['sale_year'].unique(), rotation=45)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### Average sold prices & price change % for each year

```python
average_price_yearly = property_df.groupby(['community', 'sale_year'])['sold_price'].mean().unstack()

price_change_yearly = property_df.groupby(['community', 'sale_year']).agg({
    'sold_price': lambda x: (x.iloc[-1] - x.iloc[0]) / x.iloc[0] * 100 
}).unstack()

average_price_yearly.columns = [f'avg_price_{year}' for year in average_price_yearly.columns]
price_change_yearly.columns = [f'price_change_{year}' for year in price_change_yearly.columns]

result_df = average_price_yearly.merge(price_change_yearly, on='community')

result_df.head()
```

### Change in average price by neighborhood group

```python
neighborhood_groups = {
    'South Side': 'blue',
    'Southwest Side': 'red',
    'Far Southwest Side': 'green',
    'Southeast Side': 'orange',
    'Central': 'purple',
    'Far North Side': 'brown',
    'North Side': 'pink',
    'Northwest Side': 'gray',
    'West Side': 'cyan'
}

fig_price, axs_price = plt.subplots(figsize=(15, 8))
axs_price.set_yscale('log')

property_df['last_sold_date'] = pd.to_datetime(property_df['last_sold_date'])
property_df['Year'] = property_df['last_sold_date'].dt.year

for group_name, color in neighborhood_groups.items():
    group_data = property_df[
        (property_df['Year'] >= 2014) &
        (property_df['Year'] <= 2023) &
        (property_df['neighborhood'] == group_name)
    ]

    avg_data = group_data.groupby('Year').agg({'sold_price': 'mean'}).reset_index()

    axs_price.plot(
        avg_data['Year'], avg_data['sold_price'],
        label=f'{group_name}', color=color, linewidth=2, marker='o'
    )

axs_price.set_xlim(2014, 2023)
axs_price.set_ylim(bottom=100)

axs_price.set_xlabel('Year', fontsize=14)
axs_price.set_ylabel('Average Price (Log Scale)', fontsize=14)
axs_price.set_title('Evolution of Average Price by Community (Log Scale)', fontsize=16)

axs_price.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)

axs_price.grid(True, which="both", linestyle='--', linewidth=0.5)

plt.tight_layout(rect=[0, 0, 0.75, 1])

plt.show()
```

### Price differences in 1,2,3 BR across all property styles in each neighborhood group

```python
def analyze_property_prices(property_df, bedroom_count):
    bedroom_df = property_df[property_df['beds'] == bedroom_count]

    result_df = bedroom_df.groupby(['neighborhood', 'style'])['sold_price'].mean().reset_index()

    result_pivot = result_df.pivot(index='neighborhood', columns='style', values='sold_price')

    result_pivot.fillna(0, inplace=True)

    return result_pivot

bedroom_counts = [1, 2, 3]

for bedroom_count in bedroom_counts:
    result_pivot = analyze_property_prices(property_df, bedroom_count)

    result_pivot.plot(kind='bar', figsize=(12, 6))
    plt.title(f"Average Prices of {bedroom_count}-Bedroom Properties by Property Style")
    plt.xlabel("Neighborhood")
    plt.ylabel("Average Price")
    plt.legend(title="Property Style")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## Market activity

### Most active markets

```python
property_df['last_sold_date'] = pd.to_datetime(property_df['last_sold_date'])

grouped_data = property_df.groupby('community')

sold_price_count = grouped_data['sold_price'].count()
mean_days_on_mls = grouped_data['days_on_mls'].mean()

property_df['year'] = property_df['last_sold_date'].dt.year
median_price_yearly = property_df.groupby(['community', 'year'])['sold_price'].median().unstack()

price_appreciation = median_price_yearly.pct_change(axis=1).mean(axis=1)

analysis_results = pd.DataFrame({
    'Sold Price Count': sold_price_count,
    'Mean Days on MLS': mean_days_on_mls,
    'Price Appreciation': price_appreciation
}).reset_index()

analysis_results.head()
```

### Average days on market by neighborhood by year

```python
property_df['last_sold_date'] = pd.to_datetime(property_df['last_sold_date'])

fig_dom, axs_dom = plt.subplots(figsize=(15, 8))

unique_neighborhoods = property_df['neighborhood'].unique()

for neighborhood in unique_neighborhoods:
    group_data = property_df[
        (property_df['last_sold_date'].dt.year >= 2014) &
        (property_df['last_sold_date'].dt.year <= 2023) &
        (property_df['neighborhood'] == neighborhood)
    ]

    group_data['Year'] = group_data['last_sold_date'].dt.year

    avg_dom_data = group_data.groupby('Year').agg({'days_on_mls': 'mean'}).reset_index()

    axs_dom.plot(
        avg_dom_data['Year'], avg_dom_data['days_on_mls'],
        label=f'{neighborhood}', linewidth=2, marker='o'
    )

axs_dom.set_xlim(2014, 2023)
axs_dom.set_xlabel('Year', fontsize=14)
axs_dom.set_ylabel('Average Days on Market', fontsize=14)
axs_dom.set_title('Average Days on Market by Neighborhood', fontsize=16)

axs_dom.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)
axs_dom.grid(True, which="both", linestyle='--', linewidth=0.5)

plt.tight_layout(rect=[0, 0, 0.75, 1])
plt.show()
```
